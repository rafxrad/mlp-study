```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Gerando dados de treinamento
x_train = np.linspace(0, 2*np.pi, 1000)
y_train = np.sin(x_train)

# Definindo a arquitetura da rede neural
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])

# Compilando o modelo
model.compile(optimizer='adam', loss='mse')

# Treinando o modelo
history = model.fit(x_train, y_train, epochs=100, verbose=1)

# Gerando dados para plotagem
x_test = np.linspace(0, 2*np.pi, 100)
y_pred = model.predict(x_test)

# Plotando os resultados
plt.plot(x_train, y_train, label='Dados de Treinamento')
plt.plot(x_test, y_pred, label='Predição do Modelo')
plt.xlabel('x')
plt.ylabel('sen(x)')
plt.title('Modelo MLP para sen(x)')
plt.legend()
plt.show()
```
